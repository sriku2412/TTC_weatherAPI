```python
resource_id = '8a89caa9-511c-4568-af89-7f2174b4378c'
limit = 945
language = 'English'
offset = 0

api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}&limit={}&offset={}&language={}'.format(resource_id,limit,offset,language)

print(api_url)
```

```python
COVID_19_vaccine_data.drop(columns=['_id'],inplace=True)
```

```python
columns = ["previous_day_3doses","total_individuals_at_least_one","total_individuals_partially_vaccinated","total_doses_in_fully_vaccinated_individuals","total_individuals_fully_vaccinated","total_individuals_3doses"]

for column in columns:
    COVID_19_vaccine_data[column] = pd.to_numeric(COVID_19_vaccine_data[column], errors='coerce')


COVID_19_vaccine_data['report_date'] = pd.to_datetime(COVID_19_vaccine_data['report_date'])
```
