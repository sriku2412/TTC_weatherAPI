```python


# Map OCC_MONTH to corresponding seasons
season_mapping = {
    'January': 'Winter', 'February': 'Winter', 'December': 'Winter',
    'March': 'Spring', 'April': 'Spring', 'May': 'Spring',
    'June': 'Summer', 'July': 'Summer', 'August': 'Summer',
    'September': 'Fall', 'October': 'Fall', 'November': 'Fall'
}

df['Season'] = df['OCC_MONTH'].map(season_mapping)

# Group the data by 'Season' and calculate the sum of 'sum' values
df_grouped = df.groupby('Season')['sum'].sum().reset_index()

# Define the order of the seasons
season_order = ['Winter', 'Spring', 'Summer', 'Fall']

plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='sum', data=df_grouped, palette='husl', order=season_order)
plt.xlabel('Season')
plt.ylabel('Sum of Occurrences')
plt.title('Sum of Occurrences by Season')
plt.show()
```

```python
def remove_outliers(data, threshold=3):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    filtered_data = [x for x, z in zip(data, z_scores) if abs(z) <= threshold]
    return filtered_data

filtered_data = remove_outliers(list1) #Not sure if this makes sense, can get rid of this
print(filtered_data)
```
