```python
import xml.etree.ElementTree as ET

# Specify the file path
file_path = r"C:\Users\srika\OneDrive\Documents\York\Sem-1 york\MMAI 5100 - database fundamentals\T_WHCN13_C_CWTO_202308042106_1937665482.cap"

# Load and parse the XML content from the file
with open(file_path, 'r', encoding='utf-8') as file:
    xml_content = file.read()

# Parse the XML content
root = ET.fromstring(xml_content)

# Namespace dictionary for the alert namespace
namespace = {'cap': 'urn:oasis:names:tc:emergency:cap:1.2'}

# Initialize a dictionary to store alerts based on identifiers
alerts_by_identifier = {}

# Find all alert elements
alert_elements = root.findall('.//cap:alert', namespaces=namespace)

# Loop through each alert element to separate alerts based on identifiers
for alert_element in alert_elements:
    # Extract identifier
    identifier_element = alert_element.find('.//cap:identifier', namespaces=namespace)
    if identifier_element is not None:
        identifier = identifier_element.text.strip()
        if identifier not in alerts_by_identifier:
            alerts_by_identifier[identifier] = []

        # Extract alert data (similar to the previous code)
        alert = {}
        alert_name_element = alert_element.find('.//cap:parameter/cap:valueName[.="layer:EC-MSC-SMC:1.0:Alert_Name"]/../cap:value', namespaces=namespace)
        if alert_name_element is not None:
            alert['name'] = alert_name_element.text.strip()
        polygon_element = alert_element.find('.//cap:polygon', namespaces=namespace)
        if polygon_element is not None:
            polygon_coordinates = polygon_element.text.strip().split(' ')
            alert['polygon_coordinates'] = polygon_coordinates

        # Append the alert data to the corresponding identifier
        alerts_by_identifier[identifier].append(alert)

# Print the separated alerts based on identifiers
for identifier, alerts in alerts_by_identifier.items():
    print(f"Identifier: {identifier}")
    for index, alert in enumerate(alerts, start=1):
        print(f"  Alert {index}:")
        print(f"    Alert Name: {alert.get('name', 'N/A')}")
        print("    Polygon Coordinates:")
        for coordinate in alert.get('polygon_coordinates', []):
            lat, lon = coordinate.split(',')
            print(f"      Latitude: {lat}, Longitude: {lon}")
        print("\n")
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
import pika
import ssl

# AMQP server details
amqp_server = 'dd.weather.gc.ca'
amqp_port = 5671
amqp_user = 'anonymous'
amqp_password = 'anonymous'
exchange_name = 'product.exchange'  # Replace with the actual exchange name

# Create connection parameters
credentials = pika.PlainCredentials(amqp_user, amqp_password)
context = ssl.SSLContext(ssl.PROTOCOL_TLS)
context.verify_mode = ssl.CERT_NONE  # Adjust this based on your security needs
ssl_options = pika.SSLOptions(context)
parameters = pika.ConnectionParameters(
    host=amqp_server,
    port=amqp_port,
    credentials=credentials,
    ssl_options=ssl_options
)

# Connect to the AMQP server
connection = pika.BlockingConnection(parameters)
channel = connection.channel()

# Code to consume messages from the exchange
def callback(ch, method, properties, body):
    # Parse the XML content
    xml_content = body.decode("utf-8")
    root = ET.fromstring(xml_content)

    # Initialize lists to store extracted data
    identifiers = []
    alert_names = []
    polygon_coordinates = []

    # Iterate through each 'alert' element in the XML
    for alert in root.findall(".//{urn:oasis:names:tc:emergency:cap:1.2}alert"):
        identifier = alert.find("{urn:oasis:names:tc:emergency:cap:1.2}identifier").text
        alert_name = alert.find(".//{urn:oasis:names:tc:emergency:cap:1.2}Alert_Name").text
        polygon = alert.find(".//{urn:oasis:names:tc:emergency:cap:1.2}polygon").text

        identifiers.append(identifier)
        alert_names.append(alert_name)
        polygon_coordinates.append(polygon)

    # Create a DataFrame from the extracted data
    data = {
        "Identifier": identifiers,
        "Alert Name": alert_names,
        "Polygon Coordinates": polygon_coordinates
    }

    df = pd.DataFrame(data)

    # Display the DataFrame
    print(df)

# Start consuming messages
channel.basic_consume(
    queue=exchange_name, on_message_callback=callback, auto_ack=True
)

# Keep the script running to consume messages
print('Waiting for messages. To exit press CTRL+C')
channel.start_consuming()
```
