#### 7.  Write a function to request the data<br>

```python
# imported requests to get response.
api_response = requests.get(api_url)
api_response
```

```python
data = api_response.json()
data
```

```python
approved_textbooks = pd.DataFrame(data['result']['records'])
approved_textbooks.columns
```

```python
approved_textbooks.describe(include ='all')
```

```python
approved_textbooks = approved_textbooks[['Subject', 'Title', 'Publisher', 'ISBN 13','Approval date', 'Date moving to Purchase Period Expired list']]
```

```python
new_name = {
    'ISBN 13': 'ISBN',
    'Date moving to Purchase Period Expired list': 'Expiry Date'}
approved_textbooks = approved_textbooks.rename(columns=new_name)
```

#### 8.  Illustrate your Data Cleaning and Feature Engineering processes<br>

```python
# list of books to be ordered. considering all the books are in store we are look at the changes made in the last few years
approved_textbooks['Approval date']= pd.to_datetime(approved_textbooks['Approval date'])
approved_textbooks['Expiry Date']= pd.to_datetime(approved_textbooks['Expiry Date'])
```

```python
order_textbooks = approved_textbooks[(approved_textbooks['Approval date'] >= '2022-01-01') & (approved_textbooks['Approval date'] <= '2023-12-31')]
order_textbooks = order_textbooks[['Subject', 'Title', 'Publisher', 'ISBN', 'Approval date']]
order_textbooks['Approval date'].fillna(0,inplace=True)
order_textbooks.head()
```

```python
cancel_textbooks = approved_textbooks[(approved_textbooks['Expiry Date'] >= '2022-01-01') & (approved_textbooks['Expiry Date'] <= '2023-12-31')]
cancel_textbooks = cancel_textbooks[['Subject', 'Title', 'Publisher', 'ISBN', 'Expiry Date']]
cancel_textbooks['Expiry Date'].fillna(0,inplace=True)
cancel_textbooks.head()
```

```python
with engine.connect() as connection:
    connection.execute('CREATE SCHEMA IF NOT EXISTS bookstore_requirements;')
```

```python
order_textbooks.to_sql(
    name='order_textbooks',
    schema = 'bookstore_requirements',
    con=engine,
    if_exists='replace',
    index=False,
    method='multi',
    dtype={
        'Subject': sa.types.VARCHAR(255),
        'Title': sa.types.VARCHAR(255),
        'Publisher': sa.types.VARCHAR(255),
        'ISBN': sa.types.VARCHAR(20), 
        'Approval date': sa.types.DATE,
    }
)
```

```python
cancel_textbooks.to_sql(
    name='cancel_textbooks',
    schema = 'bookstore_requirements',
    con=engine,
    if_exists='replace',
    index=False,
    method='multi',
    dtype={
        'Subject': sa.types.VARCHAR(255),
        'Title': sa.types.VARCHAR(255),
        'Publisher': sa.types.VARCHAR(255),
        'ISBN': sa.types.VARCHAR(20), 
        'Expiry Date': sa.types.DATE,
    }
)
```
