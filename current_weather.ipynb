```python
import pandas as pd
import sqlalchemy as sa
import requests
```

```python
appid = '69b9eac66ab7d3985df833d77d77bc86'
lat = '43.6548'
lon = '-79.3883'
api_url = 'https://api.openweathermap.org/data/2.5/forecast?lat={}&lon={}&appid={}'.format(lat, lon, appid)
```

```python
api_response = requests.get(api_url)
api_response
```

```python
data = api_response.json()
data
```

```python
weather_df = pd.DataFrame(data['list'])
```

```python


# Extract 'main' dictionary values into separate columns
main_columns = ['temp', 'feels_like', 'temp_min', 'temp_max', 'pressure', 'humidity', 'temp_kf']
for column in main_columns:
    weather_df[f'main_{column}'] = weather_df['main'].apply(lambda x: x[column])

# Extract 'weather' dictionary values into separate columns
weather_columns = ['id', 'main', 'description', 'icon']
for column in weather_columns:
    weather_df[f'weather_{column}'] = weather_df['weather'].apply(lambda x: x[0][column])

# Drop the original 'main' and 'weather' columns
weather_df.drop(['main', 'weather'], axis=1, inplace=True)

weather_df
```
